```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("results.csv")

plt.figure(figsize=(12,6))
plt.bar(df["Model"], df["MAE"])
plt.xticks(rotation=45)
plt.ylabel("MAE")
plt.title("MAE Comparison of Traffic Forecasting Models")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.bar(df["Model"], df["RMSE"])
plt.xticks(rotation=45)
plt.ylabel("RMSE")
plt.title("RMSE Comparison of Traffic Forecasting Models")
plt.tight_layout()
plt.show()
```
