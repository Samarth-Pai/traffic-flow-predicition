```python
A = adj_mx

A = A + np.eye(A.shape[0])

D = np.diag(
    np.power(
        A.sum(axis=1),
        -0.5
    )
)

A_hat = D @ A @ D

print(A_hat.shape)
```

```python
import pandas as pd

df = pd.read_hdf("pems-bay.h5")

print(df.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

data_scaled = scaler.fit_transform(
    df.values
)

print(data_scaled.shape)
```

```python
import numpy as np

X = []
y = []

sequence_length = 12

for i in range(
    len(data_scaled) - sequence_length
):

    X.append(
        data_scaled[
            i:i+sequence_length
        ]
    )

    y.append(
        data_scaled[
            i+sequence_length
        ]
    )

X = np.array(X)
y = np.array(y)

print(X.shape)
print(y.shape)
```

```python
split = int(
    len(X) * 0.8
)

X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]

print(X_train.shape)
print(X_test.shape)
```

```python
import torch

X_train = torch.tensor(
    X_train,
    dtype=torch.float32
)

X_test = torch.tensor(
    X_test,
    dtype=torch.float32
)

y_train = torch.tensor(
    y_train,
    dtype=torch.float32
)

y_test = torch.tensor(
    y_test,
    dtype=torch.float32
)
```
